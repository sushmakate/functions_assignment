# Task 1

```python
#define function with parameter and default discount is 5%
def apply_discount(price,discount_percent=5):
    #percent should not exceed 60%
    if discount_percent<=60:
        #calculate price after discount
        result = price - (price* (discount_percent/100))
        return result
    else:
        print('Discount should not exceed 60%')
apply_discount(500)
```

# Task 2

```python
#define function with number input
def factorial(num):
    #condition check for negative number
    if num<0:
        print('Factorial can not calculate for negative number')
        return
    else: 
        #for 0 or 1 factorial will be 1   
        if num==0 or num==1:
            return 1
        else:
            result = num * factorial(num-1)
            return result
factorial(-3)    
```

# Task 3

```python
#lambda function for calculate gst
gst = lambda price:round(price + (0.18*price),2)
print(gst(100))

#price after discount and gst
final_price = lambda price,discount: round((price - (price*(discount/100)))*(1+0.18),2)
print(final_price(100,10))
```

# Task 4

```python
#given list of prices
prices= [100, 250, 400, 1200, 50]
#generate new price list with gst using map function
prices_with_gst = list(map(lambda price:round(price+(price*0.18),2),prices))

print(prices)
print(prices_with_gst)
```

```python
#given list of prices
prices=[100, 250, 400, 1200, 50, 2000, 850]

#get prices greater than 500
new_list_prices = list(filter(lambda prices:prices>500,prices))

#get prices less than or equal to 500
price_less = list(filter(lambda prices:prices<=500,prices))

print(new_list_prices)
print(price_less)
```

```python
#function with list input
def process_prices(prices):
    #10% discount to all prices
    discounted_prices =list(map(lambda prices:prices -prices*(0.1),prices))

    #filter out prices greater than 300
    filtered_prices = list(filter(lambda price:price>300,discounted_prices))
    
    print(discounted_prices)
    print(filtered_prices)
process_prices([100,500,900,50,750])    
```
